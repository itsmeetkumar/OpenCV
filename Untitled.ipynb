```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
#Read in the image
image = cv2.imread('demo.jpg')

#print out the type of image data its dimensions
print('Image is:', type(image), 'with dimensions:', image.shape)
```

```python
%matplotlib inline

# Display the image
plt.imshow(image)
```

```python
# Make a copy of the orignal image
image_cpy = np.copy(image)

#Change color format BGR to RGB
image_cpy = cv2.cvtColor(image_cpy, cv2.COLOR_BGR2RGB)

#Display the image
plt.imshow(image_cpy)
```

```python
# Define color selection Boundaries in RGB Values 
lower_black = np.array([0, 0, 0])
upper_black = np.array([1, 1, 1])
```

```python
#Create Mask 
mask = cv2.inRange(image_cpy, lower_black, upper_black)

#view Mask 
plt.imshow(mask, cmap='gray')
```

```python
# Mask the image toshow through 

masked_image = np.copy(image_cpy)

masked_image[mask != 0] = [0, 0, 0]

plt.imshow(masked_image)
```

```python
#Load Background 
back_image = cv2.imread('background.jpg')

#Convert into RGB
back_image = cv2.cvtColor(back_image, cv2.COLOR_BGR2RGB)

#Crop it to make a right size 340 509
crop_back = back_image[0:340, 0:509]

#mask the cropped baackground so that the elephant area is blocker
crop_back[mask == 0] = [0, 0, 0]

#Display the background
plt.imshow(crop_back)

```

```python
# add the two images
comp_img = crop_back + masked_image

plt.imshow(comp_img)
```
